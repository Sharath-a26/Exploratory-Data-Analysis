```python
import pandas as pd
train = pd.read_csv("C:/Users/HP/OneDrive - Amrita Vishwa Vidyapeetham (1)/Desktop/train.csv")
train.head()
```

```python
test = pd.read_csv("C:/Users/HP/OneDrive - Amrita Vishwa Vidyapeetham (1)/Desktop/test.csv")
test.head()
```

```python
train.drop(['User_ID'],inplace = True,axis = 1)
```

```python
test.drop(['User_ID'],axis = 1,inplace = True)
```

```python
train.append(test)
```

```python
#handling categorical features

train['Gender'] = pd.get_dummies(train['Gender'],drop_first = 1)

#second method
train['Gender'] = train['Gender'].map({"F":0,"M":1})
train.head()
```

```python
#handling categorical feature age
train['Age'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
train['Age'] = label_encoder.fit_transform(train['Age'])
train.head()
```

```python
train['City_Category'].unique()
```

```python
train['City_Category'] = train['City_Category'].map({"A":1,"B":2,"C":3})
train.head()
```

```python
train.info()
```

```python
##dropping city_category feature
train.drop(['City_Category'],axis = 1,inplace = True)
```

```python
## Missing Values
train.isnull().sum()
```

```python
train['Product_Category_2'].unique()
```

```python
train['Product_Category_2'].value_counts()
```

```python
##Replacing the missing values with the mode
train['Product_Category_2'] = train['Product_Category_2'].fillna(train['Product_Category_2'].mode()[0])
train.isnull().sum()
```

```python
train['Product_Category_3'].unique()
```

```python
train['Product_Category_3'] = train['Product_Category_3'].fillna(train['Product_Category_3'].mode()[0])
train.isnull().sum()
```

```python
train['Stay_In_Current_City_Years'].dtype
```

```python
train['Stay_In_Current_City_Years'].unique()
```

```python
train['Stay_In_Current_City_Years'] = train['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
#converting object into integers
train['Stay_In_Current_City_Years'] = train['Stay_In_Current_City_Years'].astype(int)
train.info()
```

```python
import seaborn as sns
sns.pairplot(train)
```

```python
train.drop(['Product_ID'],axis = 1,inplace = True)
```

```python
y = train['Purchase']
y.head()
train.drop(['Purchase'],axis =1,inplace = True)
X = train
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state=0)
```

```python
##Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
y_pred = y_pred.astype(int)
print(r2_score(y_test,y_pred))
```

```python
import xgboost
regressor = xgboost.XGBRegressor()
regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test)
print(r2_score(y_test,y_pred))
```
